```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/KaushikBanerjee2023/A2K2HackathonTeam/main/scaled_data.csv")
data.head(10)
```

```python
for col in data.columns:
  print(col + ':' + str(data[col].value_counts().count()))
```

```python
x_data = data.drop(columns = ['Diabetic'])
y_data = data.loc[:, ['Diabetic']]
x_data.head(2)
```

```python
y_data.head(2)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x_data, y_data, test_size = 0.3, random_state = 123)
xtrain_stats = xtrain.describe().T
xtrain_stats
```

```python
#xtrain_norm = (xtrain - xtrain_stats['min']) / (xtrain_stats['max'] - xtrain_stats['min'])
#xtrain_norm.head(5)
xtrain_norm = xtrain
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

model1 = RandomForestClassifier(n_estimators=150, max_depth=50)
model2 = SVC()
model3 = DecisionTreeClassifier(max_depth=50)

model1.fit(xtrain_norm, ytrain)
model1_result = model1.predict(xtest)
print("Model 1 with RandomForestClassifier accuracy: " + str(accuracy_score(model1_result, ytest)))
print(classification_report(ytest, model1_result))

model2.fit(xtrain_norm, ytrain)
model2_result = model2.predict(xtest)
print("Model 2 with SupportVectorMachine accuracy: " + str(accuracy_score(model2_result, ytest)))
print(classification_report(ytest, model2_result))

model3.fit(xtrain_norm, ytrain)
model3_result = model3.predict(xtest)
print("Model 3 with DecisionTreeClassifier accuracy: " + str(accuracy_score(model3_result, ytest)))
print(classification_report(ytest, model3_result))
```
